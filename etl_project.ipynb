```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#read csvs
population = pd.read_csv("population_by_country_2020.csv")
happiness = pd.read_csv("happiness.csv")
print(population.info())
print(happiness.info())
# find nulls in population data
pop_null = population.isnull().sum()

pop_null
```

```python
#found where null values lie (not in top 20)
migrants_nulls = population
migrants_nulls['null'] = population['Migrants (net)'].isnull()
migrants_nulls = migrants_nulls.loc[migrants_nulls['null']==True]
migrants_nulls.head()
```

```python
#find nulls for happiness data
happ_null = happiness.isnull().sum()
happ_null
```

```python
#merge outer to find countries that don't merge
population = population.rename(columns={'Country (or dependency)':'Country or region'})
happiness_population = pd.merge(happiness,population,on="Country or region",how="outer")
happiness_population
#locate unmerged countries
non_merged_pop = happiness_population.loc[happiness_population["Population (2020)"].isnull()]
non_merged_pop
```

```python
#see what they are named in population
pd.set_option("display.max_rows", None, "display.max_columns", None)
population
```

```python
# change Czech Republic (Czechia) to "Czech Republic" on population
#Trinidad and Tobago to Trinidad & Tobago on population
#change Côte d'Ivoire to Ivory Coast in population
#change DR Congo to Congo Kinshasa in population
#change Congo to Congo Brazzaville in population
#change State of Palestine to Palestinian Territories in population
#change Eswatini to Swaziland in population

#Kosovo does not exist in population, so if ever merged we would rerank. Kept in table for info
#North Cyprus does not exist in population, so if ever merged we would rerank.  Kept in table for info



pd.set_option("display.max_rows", 20, "display.max_columns", None)

population.loc[population['Country or region']=="Czech Republic (Czechia)","Country or region"] = "Czech Republic"
population.loc[population['Country or region']=="Trinidad and Tobago","Country or region"] = "Trinidad & Tobago"
population.loc[population['Country or region']=="Côte d'Ivoire","Country or region"] = "Ivory Coast"
population.loc[population['Country or region']=="DR Congo","Country or region"] = "Congo Kinshasa"
population.loc[population['Country or region']=="Congo","Country or region"] = "Congo Brazzaville"
population.loc[population['Country or region']=="State of Palestine","Country or region"] = "Palestinian Territories"
population.loc[population['Country or region']=="Eswatini","Country or region"] = "Swaziland"

```

```python
#change names of columns for database insertion
happiness = happiness.rename(columns={"Overall rank":'rank',"Country or region":'country',"Score":'happiness_score',
                                 "GDP per capita":"gdp_per_capita","Social support":"social_support",
                                      "Healthy life expectancy":"healthy_life_expectancy",
                                 "Freedom to make life choices":"freedom_life_choices","Generosity":"generosity",
                                    "Perceptions of corruption":"perception_of_corruption"})

population = population.rename(columns={"Country or region":'country',"Population (2020)":"population",
                                        "Yearly Change":"yearly_change","Net Change":"net_change",
                                        'Density (P/Km²)':'density_per_km_sq',"Land Area (Km²)":"land_area_km_sq",
                                        "Migrants (net)":"net_migrants","Fert. Rate":"fertility_rate",
                                        "Med. Age":"median_age","Urban Pop %":"urban_pop_perc","World Share":"world_share"})
population = population.drop(columns=["null"])
population = population.set_index("country")
population
```

```python
happiness = happiness.set_index('country')
happiness
```

```python
#create database connection
connection_string = "postgres:postgres@localhost:5432/etl_challenge"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
happiness.to_sql(name='happiness', con=engine, if_exists='append', index=True)
population.to_sql(name='population', con=engine, if_exists='append', index=True)
```

```python
engine.table_names()
```

```python
connection = engine.connect()
population_filter = pd.read_sql("SELECT country,population,yearly_change,net_migrants,median_age,urban_pop_perc FROM population", connection)
population_filter
```

```python
happiness_filter = pd.read_sql("SELECT country,rank,happiness_score FROM happiness", connection)
happiness_filter
```

```python
happiness_population_df = pd.merge(happiness_filter,population_filter,on="country")
happiness_population_df
```

```python
top_20_happy = happiness_population_df.loc[happiness_population_df["rank"] <= 20]
top_20_happy
```
